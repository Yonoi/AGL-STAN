# Process the LA adj

```python
import numpy as np
import pandas as pd
```

```python
dst_df = pd.read_csv('distance_neighborhood_la.csv')
columns = dst_df.columns
num_regions = 113
```

```python
sorted(dst_df['neighborhood-2'].unique(), key=lambda x : x)
```

```python
dst_np = np.zeros((num_regions, num_regions))
dst_np[:] = np.inf
```

```python
for index in dst_df.index:
    from_id = dst_df[columns[0]][index]
    to_id = dst_df[columns[1]][index]
    dst_np[from_id, to_id] = dst_df[columns[2]][index]
```

```python
distances = dst_np[~np.isinf(dst_np)].flatten()
std = distances.std()

dst_np = np.exp(- np.square(dst_np / std))
dst_np[dst_np < 0.2] = 0
```

```python
np.save('adj_mat.npy', dst_np)
```
